```python
#i want the names that start with keith, which is similar to 'like' in sql
#By adding case=False to your str.contains() call, you ensure that variations in capitalization are handled, allowing your filter to match 
   #"Keith", "keith", "Kieth", and similar variations.

bios[bios['name'].str.contains('Keith')]

#bios[bios['name'].str.contains('Keith', case=False)]
```

```python
#lets say im worried fee some Keith and some keith so put case = false to make it case-INsensitive which basically means i want anything 
#with these letters regardless capital or not and hina since all have K capital but i searched for keith til3ooli bardo 3aadi


# | is a reg-ex syntax which basically means or 

#This regular expression matches any string containing 'keith' or 'patrick'.
#The | symbol acts as an OR operator in regular expressions.

#NOTE: Most likely, you intend to match names that contain either "keith" or "patrick" regardless of spaces. Therefore, you should use the 
#pattern without spaces around the pipe operator. This ensures that you are correctly matching names that contain either "keith" or "patrick" 
#in any case without inadvertently including spaces in the pattern.

bios[bios['name'].str.contains('keith|patrick', case = False)].head()


#can also use or kida with isin()
#bios[bios['born_city'].isin(['London', 'Burnley'])].tail()
```

```python
#find atheltes born in cities that start with a vowel :

#^: In regex, the caret symbol ^ is used to match the start of a string.
#na=False: Ensures that NaN values are treated as False and thus excluded from the result.

bios[bios['born_city'].str.contains(r'^[aeiouAEIOU]', na= False)]
```

```python
#find atheltes whos names dont have any vowels :

#^ at the beginning of the regex asserts the position at the start of the string. It indicates that the pattern should match from the beginning of the string.
#[^aeiouAEIOU] is a character class that matches any single character not listed within the brackets. Here, [^aeiouAEIOU] matches any character that is not a vowel (both lowercase and uppercase).
# * specifies that the preceding character class [^aeiouAEIOU] can occur zero or more times. So, this means the pattern will match strings composed of any number of non-vowel characters.
#$ at the end of the regex asserts the position at the end of the string. It indicates that the pattern should match up to the end of the string.

bios[bios['name'].str.contains(r'^[^aeiouAEIOU]*$', na= False)]
```

```python
# find athletes with names ending in son or sen

#In regular expressions, the dollar sign ($) is used to denote the end of a string. 
#It ensures that the pattern matches only if it appears at the very end of the string.

bios[bios['name'].str.contains(r'son$|sen$', na= False)]
```

```python
#atheletes born in 1900s 

bios[bios['born_date'].str.contains(r'^19', na= False)]
```

```python
# threeOrMore_vowels in name 
bios[bios['name'].str.contains(r'([aeiouAEIOU].*){3,}', na= False)]
```
